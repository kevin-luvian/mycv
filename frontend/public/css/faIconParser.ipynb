```python
# Define a filename.
filename = "fontawesome-all.css"
wFilename = "icons.txt"
```

```python
def sanitize(astr):
    index = astr.find("fa-")
    astr = astr[index:]
    astr = astr.replace(":before","")
    astr = astr.replace("\n","")
    astr = astr.replace(",","")
    return astr
```

```python
def sanitize_second_arr(astr):
    res = []
    for arrStr in astr.split("."):
        if("fa-" in arrStr):
            arrStr = sanitize(arrStr)
            if(arrStr != ""):
                res.append(arrStr)
    return res
```

```python
def sanitize_all(astr):
    res = []
    for arrStr in astr.split(" "):
        if(".fa-" in arrStr):
            res += sanitize_second_arr(arrStr)
    return res
```

```python
arrays = []

with open(filename, "r") as file:
    for line in file:
        if(".fa-" in line):        
            arrays += sanitize_all(line)
    file.close()
    
for astr in arrays:
    print(astr)
```

```python
arrays.sort()
with open(wFilename, "w") as file:
    for astr in arrays:
        file.write('"'+astr+'", ')
    file.close()
```
